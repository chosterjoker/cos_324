## Part 2: 3-D visualization

You may have often seen 3-D plots of (x, y, z) values. `numpy` allows you to create these very easily, aided with nice shading to make them easier to visualize. It also gives ways to rotate the 3-D plot for better understanding.

In this problem we give you the code to generate the (x, y, z) values by calling the function funky().

(Aside: The z values correspond to the difference of two Gaussian densities, but you don't need to understand that to do the homework.)

````python
# HELPER CODE: DO NOT EDIT!
def funky(delta=0.01):
    delta = 0.01
    x = np.arange(-2.0, 2.5, delta)
    y = np.arange(-1.5, 2.5, delta)
    X, Y = np.meshgrid(x, y)
    Z1 = np.exp(-X**2 - Y**2)
    Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
    Z = Z1 - Z2
    return X, Y, Z
````

**Question 1:**
 What does the line `x = np.arange(-2.0, 2.5, delta)` do? You can consult [the documentation](https://numpy.org/doc/stable/reference/generated/numpy.arange.html) for NumPy's `arange` function.

This arange function returns evenly spaced values from -2.0 to 2.5 where the spacing is delta = 0.01.


**Question 2:** What are the dimensions of the outputs X and Y of `np.meshgrid(x, y)`? Include a brief (1-2 sentence) description of how the values of X and Y relate to the values of x and y. You can consult [the documentation](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html) for NumPy's `meshgrid` function.


x has 450 values, y has 400. X and Y both have a dimension of 450 by 400. The value of X is the x array repeated by the length of the y array. It is represented by length of y rows of the x array. The value of Y is the y array into the columns of a grid with length of x columns.

Call funky() to get (x,y,z) values of the function. Then, use `plt` to create a 3D plot of this function. For better visualization it is customary to use a heatmap with a cooler color for more negative values and warmer color for positive. Use `cm` (color map) for this.

````python
X, Y, Z = funky()

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X, Y, Z, cmap = "plasma")
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

ax.view_init(elev=30, azim=45)
plt.show()
````
